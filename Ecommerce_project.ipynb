```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#
data = {
    "Product": ["Laptop", "Mobile", "Tablet", "Headphones", "Smartwatch"],
    "Price": [50000, 20000, 15000, 2000, 5000],
    "Quantity": [5, 10, 7, 20, 8]
}

df = pd.DataFrame(data)

#To  Add Columns
df["Total_Sales"] = np.array(df["Price"]) * np.array(df["Quantity"])
df["Discount"] = df["Price"] * 0.10

# Step 3: Averages
avg_price = df["Price"].mean()
avg_sales = df["Total_Sales"].mean()

print("Average Price:", avg_price)
print("Average Sales:", avg_sales)

# Create Dashboard
fig, aux = plt.subplots(2, 2, figsize=(12, 8))

# Bar Chart - Total Sales
aux[0,0].bar(df["Product"], df["Total_Sales"],color = 'blue')
aux[0,0].set_title("Total Sales")
aux[0,0].tick_params(axis='x', rotation=30)
aux[0,0].set_xlabel("Product")
aux[0,0].set_ylabel("Sales")

# Line Chart - Price Trend
aux[0,1].plot(df["Product"], df["Price"], marker='o',color = 'green')
aux[0,1].set_title("Price Trend")
aux[0,1].set_xlabel("Product")
aux[0,1].set_ylabel("Price")

#  Pie Chart - Quantity Distribution
aux[1,0].pie(df["Quantity"], labels=df["Product"], autopct='%1.1f%%')
aux[1,0].set_title("Quantity Distribution")

# Histogram - Price Distribution
aux[1, 1].hist(df["Price"],color = 'red')
aux[1, 1].set_title("Price Distribution")

# Dashboard Title
fig.suptitle("E-Commerce Dashboard", fontsize=16)

plt.tight_layout()
plt.show()
```
